# Visualization of O-information analysis  using Hyperplot 

```python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import hyperplot
import analysis

%load_ext autoreload
%autoreload 2
```

```python
DATASET_DIR = Path.cwd() / 'data'
SAVE_DIR = Path.cwd() / 'figs'
```

# Empathy

```python
fpath = DATASET_DIR / 'Briganti2017.mat'
data = analysis.load_empathy_dataset(fpath)
```

```python
dataset = 'Empathy'
savefig = False

analysis.plot_polygons(data, internode_dists=[1.6, None], show_nodelabels=True, **{'node_size':0.035})
plt.suptitle(f'{dataset.upper()} Dataset', fontsize=20)
if savefig:
    plt.savefig(SAVE_DIR / f"{dataset}_polygons.png", dpi=300)
    
analysis.plot_areas(data)
plt.suptitle(f"{dataset.upper()} Dataset", fontsize=20)
if savefig:
    plt.savefig(SAVE_DIR / f"{dataset}_areas.png", dpi=300)
```

# Eating Disorders

```python
fpath = DATASET_DIR / 'EatingDisorders.mat'
data = analysis.load_eating_dataset(fpath)
```

```python
dataset = 'Eating Disorders'
savefig = False

analysis.plot_polygons(data, internode_dists=[1.6, None], show_nodelabels=True, **{'node_size':0.035})
plt.suptitle(f'{dataset.upper()} Dataset', fontsize=20)
if savefig:
    plt.savefig(SAVE_DIR / f"{dataset}_polygons.png", dpi=300)
    
analysis.plot_areas(data)
plt.suptitle(f"{dataset.upper()} Dataset", fontsize=20)
if savefig:
    plt.savefig(SAVE_DIR / f"{dataset}_areas.png", dpi=300)
          
analysis.plot_two_rows(data, column_spacing=2.5, nodesize=0.11, subplot_width=20, subplot_height=4)
plt.suptitle(f'{dataset.upper()} Dataset', fontsize=20)
plt.subplots_adjust(top=0.9)
if savefig:
    plt.savefig(SAVE_DIR / f"{dataset}_two-rows.png", dpi=300)

# analysis.plot_planar(data)
# plt.suptitle(f'{dataset.upper()} Dataset', fontsize=20)
# if savefig:
#     plt.savefig(SAVE_DIR / f"{dataset}_planar.png", dpi=300)
```
